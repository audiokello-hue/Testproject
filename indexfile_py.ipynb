### File Handling - Saving Data beyond the Session

- `r` - Read(file must exist)
- `w` - Write(creates or overwrites)
- `a` - Append(adds to the end without erasing)
- `x` - Create(fails if the file already exists)


- **open()**
- **close()**


Best practice always use `with open(...)` because it automatically closes the file even if an error occurs.

```python
fruits = [
    'apple',
    'mango',
    'papaya',
    'pineapple',
    'watermelon',
    'dragon fruit',
    'grape',
    'blueberry',
    'lemon',
    'orange',
    'kiwi'
]

```

```python
"""
general structure is

        with open('filename', mode) as ALIAS:
          #if a list, loop over it and write to disk
          for line in lines:
            ALIAS.write(line)

"""

with open("fruits.csv", "w") as f:
  for fruit in fruits:
    f.write(fruit)
```

```python
with open("fruits.csv", "r") as f:
  content = f.read()

print(content)
```

```python
with open("fruits.csv", "r"):

  content = open("fruits.csv", "r").read()


print(content)
```

```python
with open("fruits.csv", "r") as fruits_file:

  content = fruits_file.read()


print(content)
```

```python
with open("fruits.csv", "w") as f:
  for fruit in fruits:
    f.write(fruit + "\n")

```

```python
with open("fruits.csv", "r") as fruits_file:

  content = fruits_file.read()


print(content)


##reading line by line(memory efficient for large files)
```

```python
with open('fruits.csv', 'r'):
  content = open('fruits.csv', 'r').read()

print(content)
```

### JSON Files - A natural fit for Dictionaries
Python's **json** module lets you save dictionaries directly to a file and reload them.

```python
students
```

```python
import json

with open("students.json", "w") as f:
  json.dump(students, f, indent=4)
```

```python
#load it back- comes back as a dictionary

with open("students.json", "r") as f:
  content = json.load(f)

content
```
