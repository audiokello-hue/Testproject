```python
# loading a data frame
import pandas as pd
df = pd.read_csv('iris.csv', index_col=0)

df.head()
```

```python
import numpy as np


rng = np.random.default_rng(42) #seed 

n = 25

study_hours = rng.uniform(1, 10, n).round(1)

#exam_score
exam_score = (40 + study_hours + 5 + rng.normal(0, 8, n)).round(1)
exam_score = np.clip(exam_score, 0, 100)

age = rng.integers(18, 45, n)


class_df = pd.DataFrame({
    'study hours':study_hours,
    'exam_score':exam_score,
    'age':age}
)

class_df.head()
```

```python
# covariance between study hours and exam score
cov_matrix = class_df[['study hours', 'exam_score']].cov()
cov_matrix
```

```python
corr_study_score = class_df['study hours'].corr(class_df['exam_score'])
corr_study_score
```

```python
corr_age_score = class_df['age'].corr(class_df['exam_score'])
corr_age_score
```

```python
class_df.corr()
```

```python
### create visulizations with python
### matplotlib - static visualizations ( Tableau, PowerBI, Excel, Google Data Studio)
### Has alot of syntax
### Seaborn - Library built on top of matplotlib, easier to use, less syntax
```

```python
import matplotlib.pyplot as plt

### scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(class_df['study hours'], class_df['exam_score'], color='red')
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(6, 4))
plt.hist(class_df['exam_score'], bins=10, edgecolor='black')
plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
# correlation matrics
import seaborn as sns
plt.figure(figsize=(6, 4))
sns.heatmap(class_df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation heatmap')
plt.show()
```
