import numpy as np


rng = np.random.default_rng(42) #seed 

n = 25

study_hours = rng.uniform(1, 10, n).round(1)

#exam_score
exam_score = (40 + study_hours + 5 + rng.normal(0, 8, n)).round(1)
exam_score = np.clip(exam_score, 0, 100)

age = rng.integers(18, 45, n)


class_df = pd.DataFrame({
    'study hours':study_hours,
    'exam_score':exam_score,
    'age':age}
)

class_df.head()

```python
# covariance between study hours and exam score
cov_matrix = class_df[['study hours', 'exam_score']].cov()
cov_matrix
```

```python
corr_study_score = class_df['study hours'].corr(class_df['exam_score'])
corr_study_score
```

```python
corr_age_score = class_df['age'].corr(class_df['exam_score'])
corr_age_score
```

```python
### Probability - PMF( probability mass function) & PDF (probability density function)

import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(seed=42) #seed

uniform_samples = rng.uniform(0, 1, size=10000) # uniform distribution
plt.figure(figsize=(6, 4))
plt.hist(uniform_samples, bins=50, edgecolor='black', color='brown')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('density')
plt.show()
```

```python
#normal distribution of heights - continuous distribution

mean, std = 170, 10
normal_samples = rng.normal(loc=mean, scale=std, size=10000)

plt.figure(figsize=(6, 4))
plt.hist(normal_samples, bins=30, edgecolor='black', color='blue', density=True, alpha=0.7, label='Normal Sample')

#pdf
x = np.linspace(normal_samples.min(), normal_samples.max(), 199)
plt.plot(x, scipy.stats.norm.pdf(x, loc=mean, scale=std), color='orange', label='sample pdf')

plt.title('Normal Distribution mean = 170, std=10')
plt.xlabel('Heights(cm)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Discrete Distribution: Flipping a coin 20 times, probability of getting heads is 0.5, we can use binomial distribution to model this scenario.   
binomial_samples = rng.binomial(n=20, p=0.5, size=10000)

plt.figure(figsize=(6, 4))
plt.hist(binomial_samples, bins=range(0, 22), edgecolor="black", color="mediumseagreen", align="left")
plt.title("Binomial Distribution (n=20 flips, p=0.5)")
plt.xlabel("Number of Heads")
plt.ylabel("Frequency")
plt.show()
```

```python
class Student:
    def __init__(self, first_name, last_name, subject):
        self.first_name = first_name
        self.last_name = last_name
        self.subject = subject

    def capitalize(self):
        fullname = self.first_name + " " + self.last_name
        return fullname.upper()
    
    def welcome(self):
        # Dear first_name welcome to Moringa School and the subject course
        sentence = f"Dear {self.first_name.title()}, welcome to Moringa School and the {self.subject.lower()} course!"
        return sentence

student_1 = Student('Audi', 'Timothy', 'data science')

student_1.capitalize()
```

```python
student_1.welcome()
```

```python
student_2 = Student('JAcoB', 'Tai', 'Software Development')
student_3 = Student('rAcheL', 'WanjIku', 'CyberSecurity')
student_4 = Student('Mercy', 'Korir', 'DevOps')

new_students = [student_1,
                student_2,
                student_3,
                student_4
]
```

```python
for new_student in new_students:
    print(new_student.welcome())
```
