### Working with REGEX
1. Search for patterns
2. Extracting information
3. Replacing text
4. Splitting strings

```python
#search for patterns

import re


example1 = 'It rained cats and dogs and my cat was very scared of the caterpillar.'

pattern = 'dogs'

match = re.search(pattern, example1)

if match:
  print('Pattern successfully found!!')
else:
  print('Pattern missing!!')
```

```python
mpesa = "UDM3ZJQ44F Confirmed.You have received Ksh50.00 from CAROLINE  ABUGA 0703992559 on 22/4/24 at 8:27 PM  New M-PESA balance is Ksh6,218.04.  Use a unique M-PESA PIN to keep your money safe - don't use your date of birth as your PIN. RDM3ZJQ43F"
```

```python
## extracting information

pattern = 'cat'

re.findall(pattern, example1)
```

```python
## replacing text

text = 'Hello, World!!'

pattern = r"World"


re.sub(pattern, "Mary", text)
```

```python
#splitting a string

text = 'apple,banana,cherry,mango,pineapple'

pattern = r','

fruits = re.split(pattern, text)

fruits
```

### Key components of Regular Expressions
1. `Literals`- normal characters that match  themselves
2. `Metacharacters` - special characters with specific meaning in REGEX eg. `, . ^ $ * + ? {} [] \ | ()`
3. `Character classes` - Sets of characters to match eg [0-9], [a-z], [A-Z]
4. `Anchors` - specify position in the text for example; ^ start or a text, $ end of the text
